What is Benford's Law?

Benford’s Law, also known as the First-Digit Law or The Law of Anomalous Numbers, is a statistical principle that predicts the frequency distribution of the first digits in many naturally occurring datasets.

Uses:-

1.   Fraud detection in financial data.
2.   Election results verification.
3.   Scientific data validation.
4.   Natural datasets analysis.
5.   Quality control in data collection.

```python
# By the following method we can get the first digit of the respective population.
df['population'] = df['population'].astype(str)
first_digit = df['population'].str[0].sort_values().value_counts()
print(first_digit)
```

```python
# By this we got the total length of the data set.
total_length = df['city'].sort_index().index[-1]
print(total_length)
```

```python
# Finally we got the Frequency distribution for the dataset, which shows the frequency of each digit in the first place.
frequency = (first_digit / total_length) * 100
print(round(frequency, 4))
```

```python
# This is the bar graph to represent the frquency of the digits in the first place.
import matplotlib.pyplot as plt


plt.bar(frequency.index, frequency.values, color='lightgreen', edgecolor='black')
plt.title("First Digit Frequency Population")
plt.xlabel("First Digit")
plt.ylabel("Frequency")
plt.xticks(range(1, 10))

plt.show()
```

```python
df['population_proper'] = df['population_proper'].astype(str)
first_digi = df['population_proper'].str[0].sort_values().value_counts()
print(first_digi)
```

```python
total_length = df['city'].sort_index().index[-1]
print(total_length)
```

```python
frequencys = (first_digi / total_length) * 100
print(round(frequencys, 4))
```

```python
import matplotlib.pyplot as plt


plt.bar(frequencys.index, frequencys.values, color='lightblue', edgecolor='black')
plt.title("First Digit Frequency Population_Proper")
plt.xlabel("First Digit")
plt.ylabel("Frequency")
plt.xticks(range(1, 10))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10)
width = 0.35

plt.bar(x - width/2, frequencys.values, width, label='Population_Proper', color='lightblue', edgecolor='black')
plt.bar(x + width/2, frequency.values, width, label='Population', color='lightgreen', edgecolor='black')

plt.title("First Digit Frequency Comparison")
plt.xlabel("First Digit")
plt.ylabel("Frequency")
plt.xticks(x)
plt.legend()

plt.show()
```

From observing both population and proper population, we conclude that they follow benford's Law and values are nearly equal.
